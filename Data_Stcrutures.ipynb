```python
 # Program to demonstrate string operations in Python

s1 = input("Enter a string: ")
s2 = input("Enter another string: ")

print("\nString Operations")
print("Length of first string        :", len(s1)) # 5
print("Length of second string       :", len(s2)) # 3

print("Concatenation                 :", s1 + " " + s2) #apple bad
print("Repetition (first string * 2) :", s1 * 2) # apple apple

print("Uppercase                     :", s1.upper()) #APPLE
print("Lowercase                     :", s1.lower()) #apple
print("Title case                    :", s1.title()) #Apple

print("String comparison             :", s1 == s2) # False

print("First character               :", s1[0]) # a
print("Last character                :", s1[-1])#e
print("Last character                :", s1[0:-1])#appl

print("Substring (first 3 chars)     :", s1[:3]) #app
print("Reverse string                :", s1[::-1]) #elppa

print("Replace 'a' with '@'          :", s1.replace('a', '@'))#
print("Count of 'a'                  :", s1.count('a')) # 1
print("Find 'a' position             :", s1.find('a')) #True
```
